```python
import os
import keras
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten

from skimage import transform
from skimage.io import imread
from skimage.color import rgb2gray
from sklearn.metrics import classification_report, confusion_matrix
```

```python
"""
    Preprocess image, normalizing and resizing it

    :param frame: RGBA frame
"""    
def preprocess_image(frame):
    
    # Normalize Pixel Values
    normalized_frame = frame/255.0 - 0.5
    
    # Resize
    preprocessed_frame = transform.resize(normalized_frame, [150,150])
    
    return preprocessed_frame
```

```python
doc_types = {}
imgs = []
img_type = []
dt = 0

# traverse root directory, and list directories as dirs and files as files
for root, dirs, files in os.walk("./data/dataset"):
    path = root.split(os.sep)
    for file in files:
        fname = file.split('.')[-1]
        if fname == 'fmt':
            f = open(root + '/' + file)
            doc_type = f.readline()[:-1]
            if doc_type not in doc_types:
                doc_types[doc_type] = dt
                dt += 1
            img_type.append(doc_types[doc_type])
            print(doc_type + " " + str(dt))
        elif fname == 'png':
            imgs.append(preprocess_image(imread(root + '/' + file)))

        
```

```python
print(imgs[1].shape)
print(dt)
```

```python
model = Sequential()
model.add(Conv2D(20, kernel_size=(7, 7), strides=(1, 1),
                 activation='relu',
                 input_shape=(150, 150, 1)))
model.add(MaxPooling2D(pool_size=(4, 4), strides=(1, 1)))
model.add(Conv2D(50, kernel_size = (5, 5), activation='relu'))
model.add(MaxPooling2D(pool_size=(4, 4)))
model.add(Flatten())
model.add(Dense(1000, activation='relu'))
model.add(Dense(1000, activation='relu'))
model.add(Dense(dt, activation='softmax'))

model.compile(loss=keras.losses.categorical_crossentropy,
              optimizer=keras.optimizers.SGD(lr=0.01),
              metrics=['accuracy'])
```

```python
np_img_type = np.zeros((len(img_type), dt))

print(len(img_type))

for i in range(len(img_type)):
    np_img_type[i, img_type[i]] = 1
print(np_img_type)
```

```python
x_train = np.expand_dims(np.asarray(imgs[:2000]), axis=3)
y_train = np_img_type[:2000,:]

x_val = np.expand_dims(np.asarray(imgs[2000:4000]), axis=3)
y_val = np_img_type[2000:4000,:]

x_test = np.expand_dims(np.asarray(imgs[4000:]), axis=3)
y_test = np_img_type[4000:,:]

print(y_train.shape)
```

```python
model.fit(x_train, y_train,
          batch_size=100,
          epochs=2,
          verbose=1,
          validation_data=(x_val, y_val))
```

```python
#Confution Matrix and Classification Report
Y_pred = model.predict_classes(x_test)
```

```python
print(Y_pred)
```

```python
y_pred = Y_pred
y_true = np.argmax(y_test, axis=1)
print('Confusion Matrix')
print(confusion_matrix(y_true, y_pred))
```
